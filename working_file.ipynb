Aggregating and Merging


```python
# aggregating the revenue
agg_revenue = revenue_df.groupby('userId').agg({'amount': 'sum'}).reset_index()
agg_revenue
```

```python
# merging the aggregated revenue with installs to get the revenue amount
revenue_installs = pd.merge(installs_df,agg_revenue, on = 'userId',how='left')
revenue_installs
```

```python
# validating
print("revenue before merging table installs",revenue_df['amount'].sum())
print("revenue after merging table installs",revenue_installs['amount'].sum())
```

```python
# aggregating the revenue + installs tables with countryCode,os_name,channel


revenue_installs =revenue_installs[['countryCode','os_name','channel','amount']]
revenue_installs = revenue_installs.groupby(['channel','countryCode','os_name']).agg({'amount':'sum'}).reset_index()
revenue_installs.sort_values(by=['channel', 'countryCode', 'os_name'])
```

```python
ad_spend_df = ad_spend_df[['channel','country_code','os_name','network_clicks','network_impressions','network_installs','network_installs_diff','installs','cost']]
ad_spend_df.sort_values(by=['channel', 'country_code', 'os_name'])
ad_spend_df = ad_spend_df.groupby(['channel','country_code','os_name']).agg({'installs': 'sum',
                                                                              'cost': 'sum',
                                                                              'network_clicks':'sum',
                                                                              'network_impressions':'sum',
                                                                              'network_installs':'sum',
                                                                              'network_installs_diff':'sum'
                                                                              }).reset_index()
ad_spend_df
```

```python
# combining all the data
combined_df = pd.merge(ad_spend_df,revenue_installs,left_on=['channel', 'country_code', 'os_name'],right_on=['channel', 'countryCode', 'os_name'],how='left')
combined_df
```

```python
combined_df.fillna(0)
# combined_df.to_csv('final_data.csv', index=False)

```

```python
# validating

print('Before merging data with revenue_installs \n',ad_spend_df[['network_clicks','network_impressions','network_installs','network_installs_diff','installs','cost']].sum())
print('After merging data with revenue_installs \n',combined_df[['network_clicks','network_impressions','network_installs','network_installs_diff','installs','cost']].sum())
```

Metics & Insights

```python
combined_df['ad_spend_efficiency'] = combined_df['amount'] / combined_df['cost']
combined_df['CPI'] = combined_df['cost'] / combined_df['installs']
combined_df['ARPU'] = combined_df['amount'] / combined_df['installs']

combined_df['CTR'] = combined_df['network_clicks'] / combined_df['network_impressions']
combined_df['CVR'] = combined_df['installs'] /  combined_df['network_clicks']
combined_df['CPC'] = combined_df['cost'] /  combined_df['network_clicks']

Performance = combined_df
Performance.replace([np.inf, -np.inf], 0, inplace=True)
Performance.fillna(0, inplace=True)
```

```python
# metrics definition

print("Total Install = ",Performance['installs'].sum())
print("Total Clicks = ",Performance['network_clicks'].sum())
print("Total Amount spent on Ads = ",Performance['cost'].sum())
print("Total Revenue = ",Performance['amount'].sum())
print("Conversion Rate = ",Performance['CVR'].sum())
print("Cost Per Install (CPI) = ",Performance['CPI'].sum())
print("Average Revenue per User ARPU) = ",Performance['ARPU'].sum())
print("Click Through Rate  (CTR) = ",Performance['CTR'].sum())
print("Cost Per Click (CPC) = ",Performance['CPC'].sum())
```
